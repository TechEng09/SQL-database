```python
mydb = mysql.connector.connect(
    host="localhost",
    username="username",
    password = "password",
    database = "working"
)

mynew = mydb.cursor()
```

```python

mynew.execute("CREATE TABLE customers (name VARCHAR(255), address VARCHAR(255))")
```

```python
sql = "INSERT INTO customers (name, address) VALUES (%s, %s)"
value = ("Awwal", "Ipaja, Lagos")
mynew.execute(sql,value)
mydb.commit()
```

```python
print(mynew.rowcount, "was inserted")
```

```python
new = "INSERT INTO customers (name, address) VALUES (%s, %s)"
old = [
  ('Peter', 'Lowstreet 4'),
  ('Amy', 'Apple st 652'),
  ('Hannah', 'Mountain 21'),
  ('Michael', 'Valley 345'),
  ('Sandy', 'Ocean blvd 2'),
  ('Betty', 'Green Grass 1'),
  ('Richard', 'Sky st 331'),
  ('Susan', 'One way 98'),
  ('Vicky', 'Yellow Garden 2'),
  ('Ben', 'Park Lane 38'),
  ('William', 'Central st 954'),
  ('Chuck', 'Main Road 989'),
  ('Viola', 'Sideway 1633')
]

mynew.executemany(new, old)
mydb.commit()
print(mynew.rowcount, "was inserted")
```

```python
table = "CREATE TABLE worked (id INT AUTO_INCREMENT PRIMARY KEY, surname VARCHAR (255), position VARCHAR (255))"
mynew.execute(table)

```

```python
insert = "INSERT INTO worked (surname, position) VALUES (%s, %s)"
val = [
    ("Adebayo", "MD"),
    ("Abdulhammed", "senior Manager"),
    ("Abubakar", "clerk"),
    ("Garuba", "Head Manager")
]
mynew.executemany(insert, val)
mydb.commit()
```

```python
join = "SELECT c.name, w.surname FROM customers AS c INNER JOIN worked AS w ON c.id = w.id"

mynew.execute(join)
myjoin = mynew.fetchall()

for t in myjoin:
    print(t)
```
